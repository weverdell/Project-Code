```python
#Import all necessary packages
import matplotlib.pyplot as plt
import xarray as xr
import numpy as np
import matplotlib.gridspec as gs
import cartopy.crs as ccrs
from cartopy.util import add_cyclic_point
import cmcrameri.cm as cmc
```

```python
#Read all files in

#CanESM5 Model
canesmtasfut = xr.open_dataset('tas_pdSST-futAntSIC_CanESM5.nc') #Future surface air temperature data (same order for all models)
canesmtaspd = xr.open_dataset('tas_pdSST-pdSIC_CanESM5.nc') #Present day surface air temperature data

canesmprfut = xr.open_dataset('pr_pdSST-futAntSIC_CanESM5.nc') #Future precipitation data
canesmprpd = xr.open_dataset('pr_pdSST-pdSIC_CanESM5.nc') #Present day precipitation data

canesmpslfut = xr.open_dataset('psl_pdSST-futAntSIC_CanESM5.nc') #Future sea level pressure data
canesmpslpd = xr.open_dataset('psl_pdSST-pdSIC_CanESM5.nc') #Present day sea level pressure data


#CESM2 Model
cesmtasfut = xr.open_dataset('tas_pdSST-futAntSIC_CESM2.nc')
cesmtaspd = xr.open_dataset('tas_pdSST-pdSIC_CESM2.nc')

cesmprfut = xr.open_dataset('pr_pdSST-futAntSIC_CESM2.nc')
cesmprpd = xr.open_dataset('pr_pdSST-pdSIC_CESM2.nc')

cesmpslfut = xr.open_dataset('psl_pdSST-futAntSIC_CESM2.nc')
cesmpslpd = xr.open_dataset('psl_pdSST-pdSIC_CESM2.nc')


#E3SM-1-0 Model
e3smtasfut = xr.open_dataset('tas_pdSST-futAntSIC_E3SM-1-0.nc')
e3smtaspd = xr.open_dataset('tas_pdSST-pdSIC_E3SM-1-0.nc')

e3smprfut = xr.open_dataset('pr_pdSST-futAntSIC_E3SM-1-0.nc')
e3smprpd = xr.open_dataset('pr_pdSST-pdSIC_E3SM-1-0.nc')

e3smpslfut = xr.open_dataset('psl_pdSST-futAntSIC_E3SM-1-0.nc')
e3smpslpd = xr.open_dataset('psl_pdSST-pdSIC_E3SM-1-0.nc')


#FGOALS-f3-L Model
fgoalstasfut = xr.open_dataset('tas_pdSST-futAntSIC_FGOALS-f3-L.nc')
fgoalstaspd = xr.open_dataset('tas_pdSST-pdSIC_FGOALS-f3-L.nc')

fgoalsprfut = xr.open_dataset('pr_pdSST-futAntSIC_FGOALS-f3-L.nc')
fgoalsprpd = xr.open_dataset('pr_pdSST-pdSIC_FGOALS-f3-L.nc')

fgoalspslfut = xr.open_dataset('psl_pdSST-futAntSIC_FGOALS-f3-L.nc')
fgoalspslpd = xr.open_dataset('psl_pdSST-pdSIC_FGOALS-f3-L.nc')


#HadGem3-GC31-MM Model
hadgemtasfut = xr.open_dataset('tas_pdSST-futAntSIC_HadGEM3-GC31-MM.nc')
hadgemtaspd = xr.open_dataset('tas_pdSST-pdSIC_HadGEM3-GC31-MM.nc')

hadgemprfut = xr.open_dataset('pr_pdSST-futAntSIC_HadGEM3-GC31-MM.nc')
hadgemprpd = xr.open_dataset('pr_pdSST-pdSIC_HadGEM3-GC31-MM.nc')

hadgempslfut = xr.open_dataset('psl_pdSST-futAntSIC_HadGEM3-GC31-MM.nc')
hadgempslpd = xr.open_dataset('psl_pdSST-pdSIC_HadGEM3-GC31-MM.nc')


#IPSL-CM6A-LR Model
ipsltasfut = xr.open_dataset('tas_pdSST-futAntSIC_IPSL-CM6A-LR.nc')
ipsltaspd = xr.open_dataset('tas_pdSST-pdSIC_IPSL-CM6A-LR.nc')

ipslprfut = xr.open_dataset('pr_pdSST-futAntSIC_IPSL-CM6A-LR.nc')
ipslprpd = xr.open_dataset('pr_pdSST-pdSIC_IPSL-CM6A-LR.nc')

ipslpslfut = xr.open_dataset('psl_pdSST-futAntSIC_IPSL-CM6A-LR.nc')
ipslpslpd = xr.open_dataset('psl_pdSST-pdSIC_IPSL-CM6A-LR.nc')


#MIROC6 Model
miroctasfut = xr.open_dataset('tas_pdSST-futAntSIC_MIROC6.nc')
miroctaspd = xr.open_dataset('tas_pdSST-pdSIC_MIROC6.nc')

mirocprfut = xr.open_dataset('pr_pdSST-futAntSIC_MIROC6.nc')
mirocprpd = xr.open_dataset('pr_pdSST-pdSIC_MIROC6.nc')

mirocpslfut = xr.open_dataset('psl_pdSST-futAntSIC_MIROC6.nc')
mirocpslpd = xr.open_dataset('psl_pdSST-pdSIC_MIROC6.nc')


#NorESM2-LM Model
nortasfut = xr.open_dataset('tas_pdSST-futAntSIC_NorESM2-LM.nc')
nortaspd = xr.open_dataset('tas_pdSST-pdSIC_NorESM2-LM.nc')

norprfut = xr.open_dataset('pr_pdSST-futAntSIC_NorESM2-LM.nc')
norprpd = xr.open_dataset('pr_pdSST-pdSIC_NorESM2-LM.nc')

norpslfut = xr.open_dataset('psl_pdSST-futAntSIC_NorESM2-LM.nc')
norpslpd = xr.open_dataset('psl_pdSST-pdSIC_NorESM2-LM.nc')

#View datasets to check units (comment out as necessary)
canesmtasfut
canesmprfut
canesmpslfut
```

```python
###Surface Air Temperature data (TAS variable) Annual Means
#Find difference between future data and past data for each model
canesmtas = canesmtasfut - canesmtaspd
cesmtas = cesmtasfut - cesmtaspd
e3smtas = e3smtasfut - e3smtaspd
fgoalstas = fgoalstasfut - fgoalstaspd
hadgemtas = hadgemtasfut - hadgemtaspd
ipsltas = ipsltasfut - ipsltaspd
miroctas = miroctasfut - miroctaspd
nortas = nortasfut - nortaspd
```

```python
#TAS Annual Mean Plots 

#Set scale between 3 and - 3 degrees in 11 intervals
levels = np.linspace(-3,3,11)
levels  = (levels + np.diff(levels)[0]/2)[:-1] #Centre the scale around 0 so 0 isn't a boundary but the centre of a level

cmap = plt.cm.RdBu_r #Set red/blue scale so blue is cold and red is warm

fig = plt.figure(figsize = (18,18)) #Set up figure and size
grid = gs.GridSpec(4,2) #Set grid of 4x2 plots within figure

#CANESM5 Model
tas_average = canesmtas.tas.mean('time') #Average the tas across whole year for one annual mean
plotting_lon = canesmtas.LON.values #Define longitude values to plot against
plotting_data = tas_average.values #Define data to plot - the tas variable
plotting_data, plotting_lon = add_cyclic_point(plotting_data, coord=plotting_lon) #Add cyclic point to account for 0 degrees longitude

ax = fig.add_subplot(grid[0,0],projection = ccrs.PlateCarree()) #subplot on gridpoint with world map projection
#plot data onto map, extending scale at end of colour bar
pt = ax.contourf(plotting_lon,canesmtas.LAT,plotting_data, levels=levels, cmap=cmap, transform=ccrs.PlateCarree(), extend = 'both') 
ax.coastlines() #add coastlines to map
ax.set_title('CanESM5') #set subplot title
colbar = plt.colorbar(pt) #add colour bar to map
colbar.set_label('Temperature Change (K)') #label colour bar
colbar.set_ticks([-3,-2,-1,0,1,2,3]) #set regular intervals along colour bar

#CESM2 Model
tas_average = cesmtas.tas.mean('time')
plotting_lon = cesmtas.LON.values 
plotting_data = tas_average.values
plotting_data, plotting_lon = add_cyclic_point(plotting_data, coord=plotting_lon)


ax = fig.add_subplot(grid[0,1],projection = ccrs.PlateCarree())
pt = ax.contourf(plotting_lon,cesmtas.LAT,plotting_data, levels=levels, cmap=cmap, transform=ccrs.PlateCarree(), extend = 'both') 
ax.coastlines()
ax.set_title('CESM2')
colbar = plt.colorbar(pt)
colbar.set_label('Temperature Change (K)')
colbar.set_ticks([-3,-2,-1,0,1,2,3])

#E3SM-1-0 Model
tas_average = e3smtas.tas.mean('time')
plotting_lon = e3smtas.LON.values 
plotting_data = tas_average.values
plotting_data, plotting_lon = add_cyclic_point(plotting_data, coord=plotting_lon)


ax = fig.add_subplot(grid[1,0],projection = ccrs.PlateCarree())
pt = ax.contourf(plotting_lon,e3smtas.LAT,plotting_data, levels=levels, cmap=cmap, transform=ccrs.PlateCarree(), extend = 'both') 
ax.coastlines()
ax.set_title('E3SM-1-0')
colbar = plt.colorbar(pt)
colbar.set_label('Temperature Change (K)')
colbar.set_ticks([-3,-2,-1,0,1,2,3])

#FGOALS-f3-L Model
tas_average = fgoalstas.tas.mean('time')
plotting_lon = fgoalstas.LON.values 
plotting_data = tas_average.values
plotting_data, plotting_lon = add_cyclic_point(plotting_data, coord=plotting_lon)


ax = fig.add_subplot(grid[1,1],projection = ccrs.PlateCarree())
pt = ax.contourf(plotting_lon,fgoalstas.LAT,plotting_data, levels=levels, cmap=cmap, transform=ccrs.PlateCarree(), extend = 'both') 
ax.coastlines()
ax.set_title('FGOALS-f3-L')
colbar = plt.colorbar(pt)
colbar.set_label('Temperature Change (K)')
colbar.set_ticks([-3,-2,-1,0,1,2,3])

#HadGEM3-GC31-MM Model
tas_average = hadgemtas.tas.mean('time')
plotting_lon = hadgemtas.LON.values
plotting_data = tas_average.values
plotting_data, plotting_lon = add_cyclic_point(plotting_data, coord=plotting_lon)


ax = fig.add_subplot(grid[2,0],projection = ccrs.PlateCarree()) 
pt = ax.contourf(plotting_lon,hadgemtas.LAT,plotting_data, levels=levels, cmap=cmap, transform=ccrs.PlateCarree(), extend = 'both') 
ax.coastlines()
ax.set_title('HadGEM3-GC31-MM')
colbar = plt.colorbar(pt)
colbar.set_label('Temperature Change (K)')
colbar.set_ticks([-3,-2,-1,0,1,2,3])

#IPSL-CM6A-LR Model
tas_average = ipsltas.tas.mean('time')
plotting_lon = ipsltas.LON.values 
plotting_data = tas_average.values
plotting_data, plotting_lon = add_cyclic_point(plotting_data, coord=plotting_lon)


ax = fig.add_subplot(grid[2,1],projection = ccrs.PlateCarree()) 
pt = ax.contourf(plotting_lon,ipsltas.LAT,plotting_data, levels=levels, cmap=cmap, transform=ccrs.PlateCarree(), extend = 'both') 
ax.coastlines()
ax.set_title('IPSL-CM6A-LR')
colbar = plt.colorbar(pt)
colbar.set_label('Temperature Change (K)')
colbar.set_ticks([-3,-2,-1,0,1,2,3])

#MIROC6 Model
tas_average = miroctas.tas.mean('time')
plotting_lon = miroctas.LON.values 
plotting_data = tas_average.values
plotting_data, plotting_lon = add_cyclic_point(plotting_data, coord=plotting_lon)


ax = fig.add_subplot(grid[3,0],projection = ccrs.PlateCarree()) 
pt = ax.contourf(plotting_lon,miroctas.LAT,plotting_data, levels=levels, cmap=cmap, transform=ccrs.PlateCarree(), extend = 'both') 
ax.coastlines()
ax.set_title('MIROC6')
colbar = plt.colorbar(pt)
colbar.set_label('Temperature Change (K)')
colbar.set_ticks([-3,-2,-1,0,1,2,3])

#NorESM2-LM Model
tas_average = nortas.tas.mean('time')
plotting_lon = nortas.LON.values 
plotting_data = tas_average.values
plotting_data, plotting_lon = add_cyclic_point(plotting_data, coord=plotting_lon)


ax = fig.add_subplot(grid[3,1],projection = ccrs.PlateCarree()) 
pt = ax.contourf(plotting_lon,nortas.LAT,plotting_data, levels=levels, cmap=cmap, transform=ccrs.PlateCarree(), extend = 'both') 
ax.coastlines()
ax.set_title('NorESM2-LM')
#ADD LINES TO BOX OFF WEDDELL SEA USED IN POSTER (comment out for report figure)
ax.plot([300,340],[-80,-80], color='green', transform=ccrs.PlateCarree())
ax.plot([300,340],[-65,-65], color='green', transform=ccrs.PlateCarree())
ax.plot([300,300],[-80,-65], color='green', transform=ccrs.PlateCarree())
ax.plot([340,340],[-80,-65], color='green', transform=ccrs.PlateCarree())
colbar = plt.colorbar(pt)
colbar.set_label('Temperature Change (K)')
colbar.set_ticks([-3,-2,-1,0,1,2,3])

#Save figure with high resolution
plt.savefig('AnnualTAS.png', dpi=400)
```
